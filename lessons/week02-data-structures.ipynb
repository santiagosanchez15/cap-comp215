## List Comprehension
Provides a compact syntax for two very common sequence-processing algorithms:  Map  and Filter

Basic syntax:

```python
# [ 2 * x for x in range(100)]
# [ somthing to happen to you variable, for every variable in a certain range ]
```

### Map Algorithm
Apply the same function to every item in another sequence (i.e., provide a "mapping" from the source sequence to the target

```python
# Problem:  compute the first 10 natural squares
[i ** 2 for i in range(11)] # called map because youre mapping inputs to an output using the same algorithm
'''
You can do the same thin for tuples, dictionary, any data structure
tuple(i ** 2 for i in range(11))
dict{i ** 2 for i in range(11)}
'''
```

### The Problem: Unstructured Data!
Notice it is just a single large string!  The real data set has 36 students, and I need to do this every week!

```python
raw_data = """
  1.                 Ali Oop scored  7/ 8 = 87%


  2.          Alison Ralison scored  8/ 8 = 100%


  3.         Ambily Piturbed scored  8/ 8 = 100%


  4.  Arshan Risnot Farquared scored  5/ 8 = 62%


  5.       Ayushma Jugernaugh scored  5/ 8 = 62%


  6.       Brayden Labaguette scored  7/ 8 = 87%
"""
```

### Goal
Turn this into structured data: a list of 2-tuples, each student's full name and their integer score.

```python
students = [student.split() for student in raw_data.split("\n") if student] #the stundets part ent to if, means that it doenest have None value
#This does filter and mapping, filtering all the None values and mapping the student.split() empty lines
records = [(rec[1: -5], rec[-4]) for rec in students]
records = [(" ".join(name), score[:1]) for name, score in records]
'''We use -index because ot everyone has the same name, but they do follow the same structure '''
#Think about the output and write one line per line, because you'll be doing more with 1 well structured line than with 20 lines, etc.
records
```
